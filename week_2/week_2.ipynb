```python
from graphviz import Digraph
from IPython.display import Image

dot = Digraph(comment='Store Checkout Flowchart')

# Terminators
dot.node('A', 'Start', shape='oval')
dot.node('M', 'End', shape='oval')

# Process steps
dot.node('B', 'Customer selects items', shape='rectangle')
dot.node('C', 'Go to checkout', shape='rectangle')
dot.node('D', 'Scan items', shape='rectangle')
dot.node('E', 'Calculate total cost', shape='rectangle')
dot.node('F', 'Apply discounts (if any)', shape='rectangle')
dot.node('G', 'Choose payment method', shape='rectangle')
dot.node('H1', 'Process card payment', shape='rectangle')
dot.node('H2', 'Process cash payment', shape='rectangle')
dot.node('H3', 'Process transfer', shape='rectangle')
dot.node('K', 'Print/send receipt', shape='rectangle')
dot.node('L', 'Give purchased items', shape='rectangle')

# Decision and retry
dot.node('I', 'Payment successful?', shape='diamond')
dot.node('J', 'Retry or cancel', shape='parallelogram')  # I/O or special step

# Edges
dot.edges(['AB', 'BC', 'CD', 'DE', 'EF', 'FG'])
dot.edge('G', 'H1', label='Card')
dot.edge('G', 'H2', label='Cash')
dot.edge('G', 'H3', label='Transfer')
dot.edge('H1', 'I')
dot.edge('H2', 'I')
dot.edge('H3', 'I')
dot.edge('I', 'K', label='Yes')
dot.edge('I', 'J', label='No')
dot.edge('K', 'L')
dot.edge('L', 'M')

# Display in Jupyter Notebook
dot.format = 'png'
dot.render('store_checkout_flowchart', view=False)
Image(filename='store_checkout_flowchart.png')
```

```python
import cmath  # Supports complex numbers
import numpy as np  # For solving cubic and quartic equations

def solve_quadratic():
    print("\nQuadratic Equation: Ax² + Bx + C = 0")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))

    discriminant = cmath.sqrt(b**2 - 4*a*c)
    root1 = (-b + discriminant) / (2*a)
    root2 = (-b - discriminant) / (2*a)

    print(f"Roots: {root1} and {root2}")

def solve_cubic():
    print("\nCubic Equation: Ax³ + Bx² + Cx + D = 0")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))

    # Use numpy to find roots
    coefficients = [a, b, c, d]
    roots = np.roots(coefficients)

    print("Roots:")
    for root in roots:
        print(root)

def solve_quartic():
    print("\nQuartic Equation: Ax⁴ + Bx³ + Cx² + Dx + E = 0")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    e = float(input("Enter E: "))

    coefficients = [a, b, c, d, e]
    roots = np.roots(coefficients)

    print("Roots:")
    for root in roots:
        print(root)

# Main program
print("Equation Solver")
print("1. Quadratic Equation (Ax² + Bx + C = 0)")
print("2. Cubic Equation (Ax³ + Bx² + Cx + D = 0)")
print("3. Quartic Equation (Ax⁴ + Bx³ + Cx² + Dx + E = 0)")

choice = input("Choose the type of equation (1/2/3): ")

if choice == "1":
    solve_quadratic()
elif choice == "2":
    solve_cubic()
elif choice == "3":
    solve_quartic()
else:
    print("Invalid choice.")
```
