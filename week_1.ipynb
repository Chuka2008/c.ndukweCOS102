```python
## PROJECT 1

def simple_interest(P, R, T):
    A = P *( 1 + (R/100)*T)
    return A
def compound_interest(P, R, n, t):
    A = P * 1 + R/n ^ n*t
    return A
def annuity(PMT, R, n, t):
    A = PMT*((1 + R/n)^nt - 1)/(R/n)
    return A
    
def main():  
    print ("WELCOME TO MY CALCULATOR")
    print ("Choose your function:")
    print ("1. Simple Interest")
    print ("2. Compound Interest")
    print ("3. Annuity Plan")

    choice = int(input("Enter choice (1,2,3):"))

    if choice == 1:
        P = float(input("Enter Principal Amount"))
        R = float(input("Enter Rate Amount: "))
        T = float(input("Enter Time(Years): "))
        print ("Simple Interest Amount:",simple_interest(P, R, T))
    
    elif choice == 2:
        P = float(input("Enter Principal Amount"))
        R = float(input("Enter Rate Amount: "))
        t = float(input("Enter Time(Years): "))
        n = float(input("Enter number of compound periods"))
        print ("Compound Interest Amount:", compound_interest(P, R, n, t))
        
    elif choice == 3:
        PMT = float(input("Enter Periodic Payment Amount:"))
        R = float(input("Enter Rate Amount(%)"))
        t = float(input("Enter Time(Years)"))
        n = float(input("Enter number of payments per year:"))
        print ("Annuity Plan Amount:", annuity(PMT, R, n, t))
        
    else:
        print ("Invalid choice")
        
main()
        
```

```python
from graphviz import Digraph

# Create a Digraph object
flowchart = Digraph()

# Nodes
flowchart.node("Start", "Start", shape="oval")
flowchart.node("Choose", "Choose Calculation Type", shape="diamond")

flowchart.node("SI", "Simple Interest\nSI = (P × R × T) / 100", shape="box")
flowchart.node("CI", "Compound Interest\nCI = P × [(1 + R/100)^T - 1]", shape="box")
flowchart.node("ANN", "Annuity Plan\nA = P × [(1 + r)^n - 1] / r", shape="box")

flowchart.node("End", "End", shape="oval")

# Edges
flowchart.edge("Start", "Choose")
flowchart.edge("Choose", "SI", label="Simple Interest")
flowchart.edge("Choose", "CI", label="Compound Interest")
flowchart.edge("Choose", "ANN", label="Annuity")
flowchart.edge("SI", "End")
flowchart.edge("CI", "End")
flowchart.edge("ANN", "End")

# Show in notebook
flowchart
```
